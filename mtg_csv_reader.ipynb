```python
import pandas as pd

standard_df = pd.read_csv('./data/cards.csv', low_memory=False)
print(standard_df.info())
```

```python
print(standard_df.value_counts('printings'))
```

```python
#Drop all rows whose names are Mountain, Island, Swamp, Forest, or Plains
standard_df = standard_df[~standard_df['name'].str.contains('Mountain|Island|Swamp|Forest|Plains')]
print(standard_df.info())
```

```python
#checked to see the difference between 'type' and 'types' columns
#print(standard_df.value_counts('type'))
#print(standard_df.value_counts('types'))

#checked subtypes but data looked good
#print(standard_df.value_counts('subtypes'))


#found lower and uppercase title letters. Changed all title letters to upper case
standard_df['type'] = standard_df['type'].str.title()
standard_df['types'] = standard_df['types'].str.title()

print(standard_df.value_counts('type'))
print(standard_df.value_counts('types'))
```

```python
#print(standard_df.value_counts('availability'))
#fetching the cards that are available in paper
standard_df = standard_df[standard_df['availability'].isin(['paper', 'mtgo, paper', 'arena, mtgo, paper', 'arena', 'arena, paper'])]
print(standard_df.value_counts('availability'))
```

```python
standard_df = standard_df[['uuid', 'number', 'name', 'setCode', 'text', 'manaCost', 'manaValue', 'colors', 'colorIdentity', 'power', 'toughness', 'type', 'types', 'subtypes', 'supertypes', 'rarity', 'artist', 'printings', 'language']]
#Extracting only the English cards from the data
standard_df = standard_df.loc[standard_df['language'] == 'English']
#export standard_df to a csv file
standard_df.to_csv('./data/all_cards.csv', index=False)
print(standard_df.info())
```

```python
#Checking to see if latest set is in the data
standard_df.loc[standard_df['setCode'] == 'MKM']
```

```python
#Accessing all sets that are in standard today and creating a new dataframe with only those sets
standard_sets = ['MID', 'VOW', 'NEO', 'SNC', 'DMU', 'BRO', 'ONE', 'MOM', 'MAT', 'WOT', 'LCI', 'MKM']
standard_df = standard_df[standard_df['setCode'].isin(standard_sets)]
print(standard_df.value_counts('setCode'))
```

```python
#Remove duplicates from the names column
standard_df = standard_df.drop_duplicates(subset=['name'])
print(standard_df.value_counts('name'))
```

```python
#Exporting the data to a new csv file
standard_df.to_csv('./data/standard_cards.csv', index=False)

#Exporting the data to a json file
standard_df.to_json('standard_cards.json', orient='records')
```
